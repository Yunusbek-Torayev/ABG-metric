```python
import numpy as np
import sympy as sp 
sp.init_printing()
import matplotlib.pyplot as plt
from matplotlib import style 
```

```python
a,q,r,θ,ϕ,M,Q,L = sp.symbols('a,q,r,θ,ϕ,M,Q,L')
```

```python
Σ = r**2 + a**2*sp.cos(θ)
Σ
```

```python
f = (1-(2*M*r*sp.sqrt(Σ))/((Σ+Q**2)**(3/2))+(Σ*Q**2)/((Σ+Q**2)**2))
f
```

```python
Eff = f*(1+L**2/r**2)
Eff
```
