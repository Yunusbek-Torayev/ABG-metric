```python
import numpy as np
import sympy as sp 
import matplotlib.pyplot as plt
from matplotlib import style 

r,θ,ϕ,M,Q,L = sp.symbols('r,θ,ϕ,M,Q,L')
```

```python
Eff = (1-2*M*r**2/((r**2+Q**2)**(3/2))+r**2*Q**2/((r**2+Q**2)**2))*(1+L**2/r**2)
Eff
```

```python
E1 = ((Eff.subs(M,1)).subs(L,5)).subs(Q,0)
E2 = ((Eff.subs(M,1)).subs(L,5)).subs(Q,0.2)
E3 = ((Eff.subs(M,1)).subs(L,5)).subs(Q,0.4)
E4 = ((Eff.subs(M,1)).subs(L,5)).subs(Q,0.6)
print(E1)
print(E2)
print(E3)
print(E4)
```
