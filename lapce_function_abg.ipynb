```python
import numpy as np
import sympy as sp 
sp.init_printing()
import matplotlib.pyplot as plt
```

```python
r,θ,ϕ,M,Q,L = sp.symbols('r,θ,ϕ,M,Q,L') 
```

```python
f = 1-2*M*r**2/((r**2+Q**2)**(3/2))+r**2*Q**2/((r**2+Q**2)**2)
f
```

```python
f1 = (f.subs(M,1)).subs(Q,0)
f2 = (f.subs(M,1)).subs(Q,0.2)
f3 = (f.subs(M,1)).subs(Q,0.3)
f4 = (f.subs(M,1)).subs(Q,0.4)
f5 = (f.subs(M,1)).subs(Q,0.5)
f6 = (f.subs(M,1)).subs(Q,0.6)
print(f1)
print(f2)
print(f3)
print(f4)
print(f5)
print(f6)
```

```python
r = np.linspace(0,5,1000)

f1 = -2*r**2/(r**2)**1.5 + 1
f2 = -2*r**2/(r**2 + 0.04)**1.5 + 0.04*r**2/(r**2 + 0.04)**2 + 1
f3 = -2*r**2/(r**2 + 0.09)**1.5 + 0.09*r**2/(r**2 + 0.09)**2 + 1
f4 = -2*r**2/(r**2 + 0.16)**1.5 + 0.16*r**2/(r**2 + 0.16)**2 + 1
f5 = -2*r**2/(r**2 + 0.25)**1.5 + 0.25*r**2/(r**2 + 0.25)**2 + 1
f6 = -2*r**2/(r**2 + 0.36)**1.5 + 0.36*r**2/(r**2 + 0.36)**2 + 1

#plt.figure(figsize=(8,5), dpi=200)

ax = plt.axes()
#ax.set(facecolor = "azure")

plt.xlabel("r/M")
plt.ylabel("f(r)")
plt.grid(True)

plt.xlim(0,4)
plt.ylim(-3,2)
plt.xticks(np.arange(0, 5, step=1))

plt.axhline(color = 'k')


plt.plot(r,f1, '--', label="Q=0", color='black', lw='1.5')
plt.plot(r,f2, '--', label="Q=0.2", color='red', lw='1.2')
plt.plot(r,f3, '--', label="Q=0.3", color='green', lw='1.2')
plt.plot(r,f4, '--', label="Q=0.4", color='blue', lw='1.2' )
plt.plot(r,f5, '--', label="Q=0.5", color='yellow', lw='1.2')
plt.plot(r,f6, '--', label="Q=0.6", color='black', lw='1.2' )

plt.legend(loc="upper right")

plt.show()
```
