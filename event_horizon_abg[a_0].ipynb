```python
import numpy as np
import sympy as sp
sp.init_printing()
import matplotlib.pyplot as plt
import scipy as sc
from matplotlib import style
```

```python
a,r,θ,ϕ,M,Q,L = sp.symbols('a,r,θ,ϕ,M,Q,L') 
```

```python
Σ = r**2 + a**2*sp.cos(θ)
Σ
```

```python
f = (1-(2*M*r*sp.sqrt(Σ))/((Σ+Q**2)**(3/2))+(Σ*Q**2)/((Σ+Q**2)**2))
f
```

```python
f1 = (((f.subs(M,1)).subs(Q,0.3)).subs(θ,sp.pi/3)).subs(a,0)
f2 = (((f.subs(M,1)).subs(Q,0.3)).subs(θ,sp.pi/3)).subs(a,0.2)
f3 = (((f.subs(M,1)).subs(Q,0.3)).subs(θ,sp.pi/3)).subs(a,0.5)
f4 = (((f.subs(M,1)).subs(Q,0.3)).subs(θ,sp.pi/3)).subs(a,0.95)

print(f1)
print(f2)
print(f3)
print(f4)
```

```python
from scipy.optimize import brentq

r=np.linspace(0,5,1000)

plt.figure(dpi=200)

f1 = 0.09*r**2/(r**2 + 0.09)**2 - 2*r*np.sqrt(r**2)/(r**2 + 0.09)**1.5 + 1
f2 = -2*r*np.sqrt(r**2 + 0.02)/(r**2 + 0.11)**1.5 + 0.09*(r**2 + 0.02)/(r**2 + 0.11)**2 + 1
f3 = -2*r*np.sqrt(r**2 + 0.125)/(r**2 + 0.215)**1.5 + 0.09*(r**2 + 0.125)/(r**2 + 0.215)**2 + 1
f4 = -2*r*np.sqrt(r**2 + 0.45125)/(r**2 + 0.54125)**1.5 + 0.09*(r**2 + 0.45125)/(r**2 + 0.54125)**2 + 1

plt.plot(r,f1, '--', color='black', lw=1.2, label="a=0")
plt.plot(r,f2, '--', color='blue', lw=1.2, label="a=0.2")
plt.plot(r,f3, '--', color='green', lw=1.2, label="a=0.5")
plt.plot(r,f4, '--', color='red', lw=1.2, label="a=0.95")


plt.xlim(0,3.2)
plt.ylim(-1.5,1)

plt.axhline(color = 'red')
#plt.axvline(r=root, color = 'r', **{'linestyle': 'dashed'})

plt.title("Ergosphere for $a > 0$")
plt.xlabel("r/M")
plt.ylabel("$g_{tt}$")

style.use('seaborn')

plt.legend(loc="upper right")

plt.grid(True)

plt.show()
```

```python
# bu yerda f(1) katta ildiz uchun

from scipy.optimize import brentq
import matplotlib.pyplot as plt

def f(r):
    return 0.09*r**2/(r**2 + 0.09)**2 - 2*r*np.sqrt(r**2)/(r**2 + 0.09)**1.5 + 1

r=np.linspace(0,5,1000)

plt.ylim(-1.5,1)
plt.plot(r, f(r))
plt.axhline(color = 'k')

root, info = brentq(f, 1.5, 2.5, full_output=True)
plt.axvline(x=root, color = 'r', **{'linestyle': 'dashed'})
print(root)
```
